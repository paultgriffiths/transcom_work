```python
from matplotlib import pyplot as plt
from mpl_toolkits.basemap import Basemap, shiftgrid, cm
import matplotlib as mpl
import pandas as pd
import pylab as P
import numpy as np
import netCDF4
from matplotlib.ticker import MultipleLocator
from scipy.interpolate import *
```

```python
mrch2br2 = 3*81+1+12.
mrair = 28.8
```

```python
filename='/Users/ptg21/projects/transcom/N48_results_regrid/N48_L60_chbr3_0_tracer4.nc'
ncfile1     = netCDF4.Dataset(filename)
lona        = ncfile1.variables['lon'][:]
lata        = ncfile1.variables['lat'][:]

hybrid_ht   = ncfile1.variables['pressure']
CH2Br2      = ncfile1.variables['Tracer4_conc'][:,:,:]
CH2Br2      = CH2Br2* mrair/mrch2br2/1e-12
nmonths, nheights, ny, nx = np.shape(CH2Br2)
```

```python

plt.figure(figsize=(12,4))
plt.clf()
plt.hold(True)

for imonth in range(0,nmonths):
    plt.subplot(1,2,1)
    plt.plot(CH2Br2[imonth, :, ny/2, nx/2], hybrid_ht)
    plt.ylim((1000,0))
    plt.subplot(1,2,2)
    plt.ylim((1000,0))
    plt.plot(CH2Br2[imonth, :, ny/2-12, nx/2-12], hybrid_ht)
```
